## Условие 2: 2 задача

### 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график.
Сделайте выводы.

```python
data = df['waterfront'].value_counts()
mpl.figure(figsize=(6, 10))
explode = [0.5, 0.3]
colors = ["0.5", "pink", "#FF00BB"]
labels = ['Есть вид', 'Нет вида']

mpl.pie(data, autopct='%1.1f%%', explode=explode, colors=colors)
mpl.legend(labels)
plt.show()
```

Дома с видом на набережную составляет 0,8 %

### 2.2 Изучите распределение этажей домов

```python
data1 = df['floors'].value_counts()

plt2.figure(figsize=(8, 5))
ax = sns.barplot(x=data1.index, y=data1)
ax.set_title('Распределение этажей домов')
ax.set_xlabel('Количество этажей')
ax.set_ylabel('Количество')

for i, v in enumerate(sorted(data1.index)):
    ax.text(i, data1[v], str(data1[v]), ha='center', va='bottom')
    
```

По гистограмме видно, что основную массу составляет 1- и 2-х этажные дома.

### 2.3 Изучите распределение состояния домов

```python
data3 = df['condition'].value_counts()

plt3.figure(figsize=(8, 5))
ox = sns.barplot(x=data3.index, y=data3)
ox.set_title('Распределение состояния домов')
ox.set_xlabel('Состояние дома')
ox.set_ylabel('Количество')

for o, p in enumerate(sorted(data3.index)):
    ox.text(o, data3[p], str(data3[p]), ha='center', va='bottom')
```

Большинство домов имеет среднее состояние

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
plt4.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
data5 = df.groupby('zipcode')['price'].mean().reset_index()

mpt.figure(figsize=(100, 50))
qx = sns.barplot(data=data5, x='zipcode', y='price')
qx.set_title('Средняя цена недвижимости по районам', fontsize=40)
qx.set_xlabel('Район (zipcode)', fontsize=30)
qx.set_ylabel('Средняя цена', fontsize=30)
qx.set_xticklabels(qx.get_xticklabels(), rotation=90, fontsize=20)
qx.set_yticks(qx.get_yticks())
qx.tick_params(axis='y', labelsize=20)
mpt.show()
```

Мы можем сравнить средние значения цены для различных районов и сделать выводы о том, как район влияет на стоимость недвижимости

```python
zx = sns.boxplot(x=df['price'], y=df['bathrooms'].astype('str'), whis=20);
zx.set_xlabel('Стоимость')
zx.set_ylabel('Количество ванных комнат')
zx.set_title('Влияние количества ванных комнат на стоимость');
```

Большее количество ванных комнат занимает большее количество площади дома, из-за чего стоимость увеличивается.

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='scatter')

plt.xlabel("Площадь участка, кв. м")
plt.ylabel("Цена дома, млн")
```

Чем больше площадь дома, тем выше его стоимость

```python
data = df['condition'].value_counts()

explode = [0.1] * len(data)  

fig, ax = plt.subplots()
ax.pie(data, autopct='%1.2f%%', pctdistance=0.8, explode=explode)

ax.legend(data.index)
ax.set_title("Распределение состояния домов")
```

По диаграмме видно, что большинство домов имеет состояние на 3.
